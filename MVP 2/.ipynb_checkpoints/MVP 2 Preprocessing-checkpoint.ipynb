```python
dates = [
    "STG_START_DATE",
    "STG_END_DATE",
    "RIG_ACCEPT_DATE",
    "RIG_RELEASE_DATE"
]

for d in dates:
    rcompl[d] = pd.to_datetime(rcompl[d], dayfirst=True)
    stmstg[d] = pd.to_datetime(stmstg[d], dayfirst=True)
```

```python
inter_vars = [
    "Rig_Duration (days)",
    "Stg_Duration (hrs)",
    "Stage_Length (meters)",
    "Production (monthly)",
]
```

```python
for i in range(0, len(rcompl)):
    start_rig = rcompl.loc[i, "RIG_ACCEPT_DATE"]
    end_rig = rcompl.loc[i, "RIG_RELEASE_DATE"]
    if(pd.isnull(start_rig) or pd.isnull(end_rig)):
        rcompl.loc[i, inter_vars[0]] = -1
    else:
        temp = (end_rig - start_rig).components.days
        if(temp < 0):
            rcompl.loc[i, inter_vars[0]] = -1
        else:
            rcompl.loc[i, inter_vars[0]] = temp
    
    
    start_stim = rcompl.loc[i, "STG_START_DATE"]
    end_stim = rcompl.loc[i, "STG_END_DATE"]
    if(pd.isnull(start_stim) or pd.isnull(end_stim)):
        rcompl.loc[i, inter_vars[1]] = -1
    else:
        temp = (end_stim - start_stim).components.hours + 24 * (end_stim - start_stim).components.days
        if(temp < 0):
            rcompl.loc[i, inter_vars[1]] = -1
        else:
            rcompl.loc[i, inter_vars[1]] = temp
```

```python
for i in range(0, len(stmstg)):
    start_rig = stmstg.loc[i, "RIG_ACCEPT_DATE"]
    end_rig = stmstg.loc[i, "RIG_RELEASE_DATE"]
    if(pd.isnull(start_rig) or pd.isnull(end_rig)):
        stmstg.loc[i, inter_vars[0]] = -1
    else:
        temp = (end_rig - start_rig).components.days
        if(temp < 0):
            stmstg.loc[i, inter_vars[0]] = -1
        else:
            stmstg.loc[i, inter_vars[0]] = temp
    
    
    start_stim = stmstg.loc[i, "STG_START_DATE"]
    end_stim = stmstg.loc[i, "STG_END_DATE"]
    if(pd.isnull(start_stim) or pd.isnull(end_stim)):
        stmstg.loc[i, inter_vars[1]] = -1
    else:
        temp = (end_stim - start_stim).components.hours + 24 * (end_stim - start_stim).components.days
        if(temp < 0):
            stmstg.loc[i, inter_vars[1]] = -1
        else:
            stmstg.loc[i, inter_vars[1]] = temp
```

```python
def parse_info(df,dropped_col):
    avg_pressure = []
    avg_hhp = []
    df = df.fillna('0')
    for i in range(len(df)):
        info = df.iloc[i,1]
        if info.find('AVG. HHP') == -1:
            avg_pressure.append(info[info.find('AVG PRESSURE'):info.find('AVG HHP')])
            avg_hhp.append(info[info.find('AVG HHP'):info.find('MAX')])
        else:
            avg_pressure.append(info[info.find('AVG PRESSURE'):info.find('AVG. HHP')])
            avg_hhp.append(info[info.find('AVG. HHP'):info.find('MAX')])
    
    df['AVG PRESSURE (PSI)'] = avg_pressure
    df['AVG HHP'] = avg_hhp

    df['AVG PRESSURE (PSI)'] = df['AVG PRESSURE (PSI)'].str.replace(',','')
    df['AVG PRESSURE (PSI)'] = df['AVG PRESSURE (PSI)'].str.extract('(\d+)',expand=False)
    df['AVG PRESSURE (PSI)'] = df['AVG PRESSURE (PSI)'].apply(lambda x: float(x))

    df['AVG HHP'] = df['AVG HHP'].str.replace(',','')
    df['AVG HHP'] = df['AVG HHP'].str.extract('(\d+)',expand=False)
    df['AVG HHP'] = df['AVG HHP'].apply(lambda x: float(x))
    
    return df.drop(columns = dropped_col)
```

```python
## I imputed all the nan entries (where no information about AVG PRESSURE or AVG HHP was parsed out)
## with the mean of that column

def imputation(df):
    temp = df[['AVG PRESSURE (PSI)', 'AVG HHP']]
    temp = temp.apply(lambda col: col.fillna(col.mean()),axis=0)
    
    rig = df['RIG_ID'].to_frame()
    
    return pd.concat([rig,temp],axis=1)
```
