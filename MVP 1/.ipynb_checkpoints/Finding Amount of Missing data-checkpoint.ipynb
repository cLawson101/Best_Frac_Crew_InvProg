# Finding the Amount of Missing Data

This notebook will determine how complete each feature is in all of the files that were presented to us. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import missingno as msno
```

```python
path = "../SHARED/"

file_1 = "cross_bu_dnc_stim_stage_detail.xlsx"
file_2 = "gcbu_dnc_recompletions.xlsx"
file_3 = "gcbu_dnc_stim_stage.xlsx"
file_4 = "monthly_production.xlsx"
```

```python
df_1 = pd.read_excel(path + file_1)
df_2 = pd.read_excel(path + file_2)
df_3 = pd.read_excel(path + file_3)
df_4= pd.read_excel(path + file_4)
```

# File 1

For each dataset, we found a matrix of what was n/a and what was complete. The shaded regions are entries while the white areas in the plot are n/a

```python
msno.matrix(df_1)
```

For each column, we determined the percentage of missing data.

```python
df_1_empty = pd.DataFrame(columns=["Feature", "% Complete"])

nans_1 = df_1.isnull().sum(axis = 0).tolist()
for feature in range(0, len(df_1.columns)):
    df_1_empty.loc[feature, "Feature"] = df_1.columns[feature]
    df_1_empty.loc[feature, "% Complete"] = 1.0 - nans_1[feature] / len(df_1.index)

df_1_empty
```

```python
mean = df_1_empty["% Complete"].mean()
median = df_1_empty["% Complete"].median()
mx = df_1_empty["% Complete"].max()
mn = df_1_empty["% Complete"].min()
skew = df_1_empty["% Complete"].skew()



print("Descriptive Stats for df_1:")
print("Mean: %.4f" % mean)
print("Median: %.4f" % median)
print("Max: %.4f" % mx)
print("Min: %.4f" % mn)
print("Skew: %.4f" % skew)
```
